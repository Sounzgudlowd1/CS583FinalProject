##### This is a markdown that's pretty much an extended comment.  Below is the set of libraries that are used for deep learning.  
When you run this module all this stuff will be imported

```python
#pandas dataframes make csv reading really easy.
import pandas as pd
#types of layers and the function to convert to categorical
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
import keras
#used to split data
from sklearn.model_selection import train_test_split
```

##### Now get the data, here data is just raw iris data from iris.csv

```python
#get data
data = pd.read_csv('iris.csv')
#get only the left two columns
X_data = data.iloc[:, :2]
#get only the label column, type
y_data = data[['type']]

print("y_data:")
print(y_data[0:5])
print("X_data")
print(X_data[0:5])
```

##### Not quite the format we need for neural nets, so one-hot-encode the y data

```python
y_data = to_categorical(y_data)
y_data[0:5]
```

##### Now do splitting into test, train and validation

```python
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size = 0.3)

X_validation, X_test, y_validation, y_test = train_test_split(X_test, y_test, test_size = 0.5)
print(len(X_train))
print(len(X_test))
print(len(X_validation))
```

##### This creates a neural network that takes a length 2 input vector (len and width in our case)
#### fully connects it to 10 hidden nodes, fully connects those to 
##### to 5 hidden nodes, then fully connects those 5 to a final set of 4--one for each output class [0,1,2,3]

```python
model = Sequential()
model.add(Dense(10, input_dim = 2, activation = 'relu'))
model.add(Dense(5, activation = 'relu'))
model.add(Dense(4, activation = 'softmax'))
model.compile(optimizer = keras.optimizers.SGD(lr = 0.01),
              loss = 'categorical_crossentropy',
              metrics = ['accuracy'])
```

##### Train it on train and validation data.  YOU MUST WITHOLD TEST DATA!!!!


```python
history = model.fit(x = X_train, y = y_train, epochs = 1000, batch_size = 50, 
          validation_data = [X_validation, y_validation])
```

##### Let's see how we did!

```python
print("Train accuracy: ")
print(model.evaluate(X_train, y_train)[1]) #returns both loss and accuracy, so this is getting the accuracy

print("Test accuracy: ")
print(model.evaluate(X_test, y_test)[1])
```

Note that the results may change run to run.  The train test split is random.  You can fix the random seed by setting the random_state attribute of the train_test_split like this:
X_validation, X_test, y_validation, y_test = train_test_split(X_test, y_test, test_size = 0.5, random_state = 7)
